```python
#IMPORTANDO O Pandas

import pandas as pd
```

```python
# Lendo a base de dados

arquivo = 'cno.csv'
df = pd.read_csv(arquivo , sep =',',header=0 , encoding = ' latin 1')
```

```python
df.head()
```

```python
df.dropna(subset=['Caixa Postal' , 'NI do responsável','CNO vinculado','Complemento' ,'CEP','Nome empresarial','Nome','Data de registro'])
```

```python
df.head()
```
